 **1.기본 Library 호출**

```python
from keras.models import Sequential
from keras.utils import np_utils 
from keras.layers.core import Dense, Activation, Dropout
from keras.layers import Conv2D, MaxPooling2D, Flatten

# 훈련 데이터셋에서 검증 데이터셋
from sklearn.model_selection import train_test_split

# 데이터 분석
import pandas as pd
import numpy as np
```

**2. 데이터 받기**

```python

train = pd.read_csv('../input/digit-recognizer/train.csv')
labels = train.iloc[:,0].values.astype('int32')

X_train = (train.iloc[:,1:].values).astype('float32')
X_test = (pd.read_csv('../input/digit-recognizer/test.csv').values).astype('float32')
```

read_csv를 통해 data를 받음.

```python
y_train = np_utils.to_categorical(labels)
```

각 값의 레이블은 to_categocial을 활용하여 범주화 시킵니다.

```python
X_train = X_train.reshape((-1,28,28,1))
X_test = X_test.reshape((-1,28,28,1))
```

input 이미지가 28x28이고 channel은 흑백 1개이므로, 위와 같이 reshape 해줌.

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.1)
```

**3.Data preprocessing**

```python

scale = np.max(X_train)
X_train /= scale
X_test /= scale

mean = np.std(X_train)
X_train -= mean
X_test -= mean
```

data scale들을 조정해준다. 

**4. Model**

```python

model = Sequential()
model.add(Conv2D(32,(3,3), activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32,(3,3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(64,(3,3), activation='relu'))
model.add(Conv2D(64,(3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

layer을 쌓아서 모델을 구성한다.

모델에서 사용하는 layer의 종류는 총 5개로
Convolution layer, MaxPooling2D,Dropout,Flatten,Dense를 사용한다. 
activation fuction으로는 2종류를 사용하는데 relu와 softmax를 사용하였다.

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_data=(X_val, y_val))

print("훈련데이터 점수 : ", model.evaluate(X_train, y_train))
print("검증데이터 점수 : ", model.evaluate(X_val, y_val))
```

컴파일에서는 총 3가지를 지정한다.
옵티마이저는 경사하강법 보다 빠른 RMSprop 옵티마이저를 사용하였다.
손실함수는 categorical_crossentropy를 사용합니다. 두 확률 분포 사이의 거리를 측정하고 확률 분포 사이의 거리르 최소화하고 최적화한다.
측정지표는 accuracy로 나타내었다.
